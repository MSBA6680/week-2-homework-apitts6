```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

df = yf.download("PH", start="2013-08-01", end="2023-08-01")

df['Returns'] = df['Close'].pct_change()

df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)

plt.figure(figsize=(14, 7))
plt.subplot(3, 1, 1)
plt.plot(df.index, df['Close'], label='Parker-Hannifin Stock Price')
plt.title('Parker-Hannifin Stock Price')
plt.ylabel('Price')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df.index, df['Returns'], label='Daily Returns')
plt.title('Parker-Hannifin Daily Returns')
plt.ylabel('Returns')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df.index, df['Volatility'], label='Annualized Volatility')
plt.title('Parker-Hannifin Annualized Volatility')
plt.xlabel('Year')
plt.ylabel('Volatility')
plt.legend()

plt.tight_layout()
plt.show()
```
